# Döngüler

## For Döngüsü

```python
for i in range(5,25,3):
    print(i)
```

```python
# iç içe döngüler

for i in range(10):
    for j in range(10):
        print(i, "*", j, "=", i*j)
```

## While Döngüsü

```python
a = 0
while a < 10:
    print(a)

    #a = a + 1
    a += 1

print("Bitti")
```

### break

```python
a = 10
while a > 0:
    print(a)

    if a == 5:
        break
    
    a -= 1
```

### continue

```python
a = 10
while a > 0:
    print(a)

    if a == 5:
        continue
    
    a -= 1
```
