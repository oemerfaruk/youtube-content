# **t-SNE** (*t-Dağıtımlı Stokastik Komşu Gömme*)

```python
import pandas as pd
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

from sklearn.datasets import load_iris

from sklearn.manifold import TSNE
```

```python
iris = load_iris()
X = iris.data
y = iris.target
```

## Model Eğitme ve Grafik

```python
tsne = TSNE(n_components=3, random_state=42)
X_tsne = tsne.fit_transform(X)
```

```python
fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')

colors = ['red', 'green', 'blue']
labels = ['setosa', 'versicolor', 'virginica']

for i, (label, color) in enumerate(zip(labels, colors)):
    ax.scatter(X_tsne[y == i, 0], X_tsne[y == i, 1], X_tsne[y == i, 2],
               c=color, label=label, s=50, alpha=0.7)

ax.set_title('t-SNE ile Iris Veri Setinin 3 Boyuta İndirgenmesi', fontsize=16)
ax.set_xlabel('t-SNE Bileşen 1', fontsize=12)
ax.set_ylabel('t-SNE Bileşen 2', fontsize=12)
ax.set_zlabel('t-SNE Bileşen 3', fontsize=12)
ax.legend()
plt.show()
```
