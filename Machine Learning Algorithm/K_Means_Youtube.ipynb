# K-Means Clustering Algoritması

## Müşteri segmentasyonu
**veri** [mall-custormers.csv kaggle](https://www.kaggle.com/datasets/shwetabh123/mall-customers)

```python
import pandas as pd
import matplotlib.pyplot as plt

from sklearn.cluster import KMeans
```

```python
df = pd.read_csv("Mall_Customers.csv")
df.head()
```

```python
X = df[['Annual Income (k$)', 'Age', 'Spending Score (1-100)']]
```

```python
plt.scatter(X['Annual Income (k$)'], X['Spending Score (1-100)'])
```

```python
kmeans = KMeans(n_clusters=5)
df['Cluster'] = kmeans.fit_predict(X)
```

```python
df.head()
```

```python
# Görselleştirme
plt.figure(figsize=(8,6))
plt.scatter(X['Annual Income (k$)'], X['Spending Score (1-100)'],
            c=df['Cluster'], cmap='viridis', s=5)
plt.xlabel("Annual Income (k$)")
plt.ylabel("Spending Score")
plt.title("K-Means Clustering ile Müşteri Segmentasyonu")
plt.show()
```
