```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
# x = m^2 - y = fiyat
x = np.array([[50],[60],[70],[80],[90]])  # girdi değişkeni (bağımsız)
y = np.array([150,180,210,240,270])       # hedef değişken (bağımlı)
```

```python
model = LinearRegression()
model.fit(x,y)
```

```python
print("Eğim: ", model.coef_[0])
print("Sabit: ", model.intercept_)
```

```python
predict = model.predict([[58]])
```

```python
print("Değeriçin tahmini fiyat: ", predict[0])
```

```python
plt.scatter(x,y, color='blue')
plt.plot(x, model.predict(x), color='red')
plt.scatter(58, predict[0], color='green')
plt.xlabel("Metrekare")
plt.ylabel("Fiyat")
plt.show()
```
