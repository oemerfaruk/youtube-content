# Hiyerarşik Kümeleme (Hierarchical Clustering) Algoritması

```python
import pandas as pd
import matplotlib.pyplot as plt

from sklearn.preprocessing import StandardScaler
from scipy.cluster.hierarchy import linkage, dendrogram, fcluster
```

```python
url = "https://raw.githubusercontent.com/uiuc-cse/data-fa14/gh-pages/data/iris.csv"
iris = pd.read_csv(url)
```

```python
X = iris.iloc[:, 0:4].values
```

```python
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

### Linkage matrix

```python
linked = linkage(X_scaled, 'ward') # method = 'ward'
```

### Dendrogram

```python
plt.figure(figsize=(15, 7))
dendrogram(linked,
           orientation='top',
           distance_sort='descending',
           show_leaf_counts=True)
plt.show()
```

```python
clusters = fcluster(linked, t=3, criterion='maxclust')
print("Küme etiketleri:", clusters)
```

```python
plt.figure(figsize=(8,6))
plt.scatter(X_scaled[:, 2], X_scaled[:, 3], c=clusters, cmap='rainbow', s=50)
plt.xlabel("Petal Length (scaled)")
plt.ylabel("Petal Width (scaled)")
plt.title("Hierarchical Clustering (Iris Dataset)")
plt.show()
```
